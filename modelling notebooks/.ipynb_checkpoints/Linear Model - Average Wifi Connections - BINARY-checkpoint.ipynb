## Average number of wifi connections

Create a table in the database containing the average number of wifi connections between quarter past the hour and quarter to the hour for each hour in each room

```python
# CREATE A TABLE WITH AVERAGE NUM OF WIFI CONNECTIONS BETWEEN QUARTER PAST AND QUARTER TO THE HOUR IN A ROOM
# create table
tname = 'AverageNumWifiConnections'
sql = "CREATE TABLE IF NOT EXISTS " + tname + " (DateTime datetime, Room VARCHAR(10), AvgNumWifiConn FLOAT, PRIMARY KEY (DateTime, Room))"
cursor.execute(sql)

# Dates and Times to be queried
Date = ['2015-11-03', '2015-11-04', '2015-11-05', '2015-11-06', '2015-11-09', '2015-11-10', '2015-11-11', '2015-11-12', '2015-11-13', '2015-11-16']
Time = ['09', '10', '11', '12', '13', '14', '15', '16', '17']

for d in Date:
    for t in Time:
        # set start date time and end datetime to query between
        start = "'" + d + " " + t + ":15:00'"
        end = "'" + d + " " + t + ":45:00'"
        # set datetime to enter into database (start of the hour being queried)
        datetime = d + " " + t + ":00:00"
        # Query B-004 between set datetime
        sql1 = "SELECT AVG(Associated) FROM WiFiLogData WHERE DateTime BETWEEN " + start + " AND " + end + " AND Room = 'B-004'"
        cursor.execute(sql1)
        res1 = cursor.fetchall()
        # insert result into table
        cursor.execute("INSERT INTO " + tname + " (DateTime, Room, AvgNumWifiConn) VALUES ('%s', '%s', %f)" % (datetime, 'B-004', res1[0][0]))
        con.commit() 
        # Query B-003 between set datetime 
        sql2 = "SELECT AVG(Associated) FROM WiFiLogData WHERE DateTime BETWEEN " + start + " AND " + end + " AND Room = 'B-003'"
        cursor.execute(sql2)
        res2 = cursor.fetchall()
        # insert result into table
        cursor.execute("INSERT INTO " + tname + " (DateTime, Room, AvgNumWifiConn) VALUES ('%s', '%s', %f)" % (datetime, 'B-003', res2[0][0]))
        con.commit() 
        # Query B-002 between set datetime
        sql3 = "SELECT AVG(Associated) FROM WiFiLogData WHERE DateTime BETWEEN " + start + " AND " + end + " AND Room = 'B-002'"
        cursor.execute(sql3)
        res3 = cursor.fetchall()
        # insert result into table
        cursor.execute("INSERT INTO " + tname + " (DateTime, Room, AvgNumWifiConn) VALUES ('%s', '%s', %f)" % (datetime, 'B-002', res3[0][0]))
        con.commit() 
```

# Analytics Base Table

Dataframe contianing all data needed for modelling

```python
# JOIN -- Analytics Bast Table (ABT)
sql = """SELECT G.DateTime, W.Room, R.Capacity, T.Module, M.NumReg, W.AvgNumWifiConn, G.PercentageEstimate, G.BinaryEstimate
        FROM DatabaseMain.AverageNumWifiConnections W, DatabaseMain.GroundTruth G, DatabaseMain.Rooms R, DatabaseMain.TimeModule T, DatabaseMain.Modules M
        WHERE W.Room = G.Room AND W.DateTime = G.DateTime AND W.Room = G.Room AND W.Room = R.Room AND W.Room = T.Room AND R.Room = W.Room AND R.Room = G.Room AND T.Room = G.Room AND T.Room = R.Room AND T.Room =  W.Room AND T.DateTime = G.DateTime AND M.ModuleName = T.Module
        ORDER BY W.Room"""
df = pd.read_sql_query(sql, con)
# df.to_csv('ABT.csv', index=False)
```

# Linear regression with package statsmodel

## Multiple linear regression
### Training a model

```python
#Import statsmodels package for training a linear regression model.
import statsmodels.formula.api as sm

# Train aka fit, a model using all features
lm = sm.ols(formula="BinaryEstimate ~  Capacity + NumReg + AvgNumWifiConn + C(Room) -1 + C(DateTime)-1 + C(Module)-1", data=df).fit()

# Print the weights learned for each feature.
print(lm.params)
```

```python
# Print the summary of the trained model.
print(lm.summary())
```

### Evaluating the model on training and test data
Split the dataset into a training set and a test set. Fit the model on the training set, evaluate both on the training and the test set. Compare the prediction error on both sets.

```python
# Take the first 3/4 as training data
df_train = df[:162]
df_train.head()
```

```python
# Take the last 1/4 rows as test data
df_test = df[162:].reset_index(drop=True)
df_test.head()
```

```python
# mix up rows
# df = df.iloc[np.random.permutation(len(df))].reset_index(drop=True)
# split into training and test
# from sklearn.cross_validation import train_test_split
# df_train, df_test = train_test_split(df, test_size = 0.2)
```

## Training on df_train

```python
# Train aka fit, a model using all features.
lm = sm.ols(formula="BinaryEstimate ~  Capacity + NumReg + AvgNumWifiConn + C(Room) -1 + C(DateTime) -1 + C(Module)-1", data=df_train).fit()

# Print the weights learned for each feature.
print(lm.params)
```

```python
# Print the detailed metrics for the trained model.
print(lm.summary())
```

## Testing on df_train

```python
# test on df_train
lm.predict(df_train)
```

```python
# Get predicted values
pred =  lm.predict(df_train)

# Bin predictions into occupied (1) or not occupied (0)
predictions_train = [0 for i in range(0, len(pred))]
for i in range(0, len(pred)):
    if pred[i] > 0.5:
        predictions_train[i] = 1
        
# look at the actual values and the predicted values beside one another
fmt = '{:<8}{:<8}{}'
print(fmt.format('', 'Actual', 'Predicted'))
for i, (act, pred) in enumerate(zip(list(df_train.BinaryEstimate), predictions_train)):
    print(fmt.format(i, act, pred))
```

```python
# Pair the actual and the predicted values
print("Actual - Predicted:\n", (df_train.BinaryEstimate - lm.predict(df_train)))
print("\n(Actual - Predicted) squared:\n", (df_train.BinaryEstimate - lm.predict(df_train))**2)

# Pair the actual and the predicted values
# USING THE BINNED (0 or 1) PREDICTIONS
print("Actual - Predicted:\n", (df_train.BinaryEstimate - predictions))
print("\n(Actual - Predicted) squared:\n", (df_train.BinaryEstimate - predictions_train)**2)
```

```python
# Print the Mean Squared Error of the model on the training set
mse = ((df_train.BinaryEstimate - lm.predict(df_train))** 2).mean()
print("\nMean Squared Error:\n", mse)

# Print the Mean Squared Error of the model on the training set
# USING THE BINNED (0 or 1) PREDICTIONS
mse = ((df_train.BinaryEstimate - predictions_train) ** 2).mean()
print("\nMean Squared Error:\n", mse)
```

```python
print("|Actual - Predicted|:\n", abs(df_train.BinaryEstimate - lm.predict(df_train)))
print("BINNED VALUES -> |Actual - Predicted|:\n", abs(df_train.BinaryEstimate - predictions_train))
```

```python
# Print the Mean Absolute Error of the model on the training set
mae = abs(df_train.BinaryEstimate - lm.predict(df_train)).mean()
print("\nMean Absolute Error on training:\n", mae)

# Print the Mean Absolute Error of the model on the training set
# USING THE BINNED (0 or 1) PREDICTIONS
mae = abs(df_train.BinaryEstimate - predictions_train).mean()
print("\nMean Absolute Error (binned) on training:\n", mae)
```

## Testing on df_test

```python
[df_test.BinaryEstimate, lm.predict(df_test)]
```

```python
# Get predicted values
pred =  lm.predict(df_test)

# Bin predictions into occupied (1) or not occupied (0)
predictions_test = [0 for i in range(0, len(pred))]
for i in range(0, len(pred)):
    if pred[i] > 0.5:
        predictions_test[i] = 1
        
# look at the actual values and the predicted values beside one another
fmt = '{:<8}{:<8}{}'
print(fmt.format('', 'Actual', 'Predicted'))
for i, (act, pred) in enumerate(zip(list(df_test.BinaryEstimate), predictions_test)):
    print(fmt.format(i, act, pred))
```

```python
accuracy = 0 
for i in range(0,len(predictions_test)):
    if (predictions_test[i] == df_test['BinaryEstimate'][i]):
        accuracy += 1

accuracy = accuracy/len(df_test)        
print('Accuracy Linear Binary: ', accuracy)
```

```python
# Print the Mean Squared Error of the model on the test set
mse = ((df_test.BinaryEstimate - lm.predict(df_test))** 2).mean()
print("\nMean Squared Error on test:\n", mse)

# Print the Mean Squared Error of the model on the test set
mse = ((df_test.BinaryEstimate - predictions_test)** 2).mean()
print("\nMean Squared Error (binned) on test:\n", mse)

# Print the Mean Absolute Error of the model on the test set
mae = abs(df_test.BinaryEstimate - lm.predict(df_test)).mean()
print("\nMean Absolute Error on test:\n", mae)

# Print the Mean Absolute Error of the model on the test set
mae = abs(df_test.BinaryEstimate - predictions_test).mean()
print("\nMean Absolute Error (binned) on test:\n", mae)
```
