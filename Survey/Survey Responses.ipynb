```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('Responses.xlsx')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
df.sort_values(['Which one of the following describes you best?', 'Which best describes your field of study?'], ascending=[True, True], inplace=True)
```

```python
for column in df.columns:
    plt.figure().suptitle(column, fontsize=20)
    df[column].value_counts().plot(kind='bar')
```

# Break results down by status

```python
status = set(df['Which one of the following describes you best?'])
status
```

```python
Employed = pd.DataFrame(df.loc[df['Which one of the following describes you best?'] == 'Employed'])
Other = pd.DataFrame(df.loc[df['Which one of the following describes you best?'] == 'Other'])  
PostGrad = pd.DataFrame(df.loc[df['Which one of the following describes you best?'] == 'Postgraduate Student'])
RecGrad = pd.DataFrame(df.loc[df['Which one of the following describes you best?'] == 'Recent graduate (within the last 2 years)'])
UnderGrad = pd.DataFrame(df.loc[df['Which one of the following describes you best?'] == 'Undergraduate Student'])
```

```python
Employed=Employed.dropna(axis=1,how='all')
```

```python
# EMPLOYED 
for column in Employed.columns:
    plt.figure().suptitle(column, fontsize=20)
    Employed[column].value_counts().plot(kind='bar')
```

```python
Other = Other.dropna(axis=1,how='all')
```

```python
# Other
for column in Other.columns:
    plt.figure().suptitle(column, fontsize=20)
    Other[column].value_counts().plot(kind='bar')
```
