```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_csv('Occupants Survey.csv')
df.head(5)
```

```python
df.columns=['Timestamp', 'studentOrEmployed', 'field', 'classAttendancePercentage', 'punctuality', 'wasLate', 'classFinish', 
            'roomEmptyStudy', 'moduleOverOthers', 'numDevicesClass', 'numDevicesAtOnce', 'laptop', 'smartphone', 'otherDevices',
            'wifiOverData', 'macAddress']
df.dtypes
```

```python
# Filter participants into CS and Non-CS students.
df_CS = df[df.field == "Computer Science and Informatics"]
df_Not_CS = df[df.field != "Computer Science and Informatics"]

# Filter participants into undergraduate and postgraduate students.
df_postgrad = df[df.studentOrEmployed == "Postgraduate Student"]
df_undergrad = df[df.studentOrEmployed == "Undergraduate Student"]
```

```python
# Reference: https://www.oreilly.com/learning/pythons-pandas-make-data-analysis-easy-and-powerful-with-a-few-lines-of-code [Accessed: 21/08/16]
df_CS['numDevicesClass'].value_counts().plot(kind="pie", title = "Number of devices connected to Wi-Fi generally during lectures/tutorials - CS students")
```

```python
df_Not_CS['numDevicesClass'].value_counts().plot(kind="pie", title= "Number of devices connected to Wi-Fi generally during lectures/tutorials - Non-CS ")
```

```python
df_CS['numDevicesAtOnce'].value_counts().plot(kind="pie", title = "Number of devices typically have connected to WiFi at one time - CS students")
```

```python
df_Not_CS['numDevicesAtOnce'].value_counts().plot(kind="pie", title = "Number of devices typically have connected to WiFi at one time - Non-CS")
```
